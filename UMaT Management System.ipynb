```python

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")


class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def enroll_course(self, course_name):
        print(f"{self.name} has enrolled in {course_name}.")


    def display_info(self):
        super().display_info()
        print(f"Student ID: {self.student_id}")


class Lecturer(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

    def teach_course(self, course_name):
        print(f"{self.name} is teaching {course_name}.")



print(" Testing Student ")
stud = Student("Godwin", 19, "UMaT-ST-01")
stud.display_info()
stud.enroll_course("Introduction to Python")

print("\n Testing Lecturer ")
lect = Lecturer("Mr. Cobbinah", 40, "UMaT-LEC-99")
lect.display_info()
lect.teach_course("Object-Oriented Programming")
```
